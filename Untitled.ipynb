```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the input file.
df = pd.read_csv("video_play.csv")

# Pie-chart.

# Filtering out the null values in Games
df = df[df['game'].notnull()]

# Grouping by Games and aggregating based on device_id to count the number of players per game.
df_video_play_properties = df.groupby('game', as_index=False)['device_id'].count().rename(columns = {'device_id':'No_of_players'})

# Counting the total number of players.
total_properties = df_video_play_properties['No_of_players'].sum()

# Calculating the percentage of each game players
df_video_play_properties['percentage_of_players'] = ((df_video_play_properties['No_of_players'] / total_properties) * 100)

# Finding the Games with the top-5 percentage of players.
df_top_10_video_play_percentage = df_video_play_properties.nlargest(columns='percentage_of_players', n = 5)

# Change the size of the figure (in inches).
plt.figure(figsize=(8,6))

# Slices: percentage_of_players.
slices = df_top_10_video_play_percentage['percentage_of_players']
# Categories: ZONE_CATEGORY.
categories = df_top_10_video_play_percentage['game']
# For different colors: https://matplotlib.org/examples/color/named_colors.html
cols = ['purple', 'red', 'green', 'orange', 'dodgerblue']

# Plotting the pie-chart.
plt.pie(slices,
        labels = categories,
        colors=cols,
        startangle = 90,
#         shadow = True,
        # To drag the slices away from the centre of the plot.
        explode = (0.040,0,0,0,0),
        # To display the percent value using Python string formatting.
        autopct = '%1.1f%%'
       )

# Title of the plot.
plt.title('Top-5 games with the highest player`s percentage\n Players per game comapred.', fontsize = 12)

# Saving the figure on disk. 'dpi' and 'quality' can be adjusted according to the required image quality.
plt.savefig('Pie_chart.jpeg', dpi = 400, quality = 100)

# Displays the plot.
plt.show()

# Clears the current figure contents.
plt.clf()
```

```python
# Bar plot.

# Importing matplotlib to plot the graphs.
import matplotlib.pyplot as plt

# Importing pandas for using pandas dataframes.
import pandas as pd

# Reading the input file.
df = pd.read_csv("video_play.csv")

# Removing the null values in games.
df = df[(df['game'].notnull())]

df = df[['login', 'game']].groupby('game', as_index = False).count().rename(columns = {'login':'number_of_players'})

df2 = df.nlargest(columns='number_of_players', n = 5)

x = df2['game']

y = df2['number_of_players']

# Change the size of the figure (in inches).
plt.figure(figsize=(17,6))

# Linewidth determines the width of the line.
plt.bar(x, y, label = 'Number of players', color = 'dodgerblue',  width = 1, align = 'center', alpha = 0.7)
# plt.bar(x2, y2, label = 'Bar 2', color = 'red',  width = 1)

# X-axis label.
plt.xlabel('Games', fontsize = 16)

# Y-axis label.
plt.ylabel('Number of players', fontsize = 16)

# Title of the plot.
plt.title('Twitch - Stream Stat - 2015-01-01', fontsize = 32)

# Grid
# plt.grid(True)
plt.grid(axis='y')

# Legend for the plot.
plt.legend()

# Saving the figure on disk. 'dpi' and 'quality' can be adjusted according to the required image quality.
plt.savefig('Bar_plot.jpeg', dpi = 400, quality = 100)

# Displays the plot.
plt.show()

# Clears the current figure contents.
plt.clf()
```
